```python
import numpy as np
```

```python
arr = np.linspace(1,10,10, dtype='int8')
```

```python
arr
```

```python
indices_cond = arr > 5
```

```python
indices_cond = arr > 5
print(indices_cond)
```

```python
print(arr[indices_cond])
```

```python
print(arr[(arr > 5) & (arr < 9)])
```

```python
arr[arr > 5] = 99
print(arr)
```
