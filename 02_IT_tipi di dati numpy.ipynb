### Importare la libreria di NumPy

```python
import numpy as np
```

# **Lista Vs. Array**

```python
#Lista
[1,2,3,4]
```

***Transformiamo la lista in un Array***

```python
arr = np.array([1,2,3,4])
arr
```

***Che Tipo di Array Abbiamo?***

```python
arr.dtype
```

**Abbiamo davvero bisogno di 64 bit per uno, due, tre e quatro?.**

```python
# Esempio di conversione da decimale a binario
print(bin(1))
print(bin(2))
print(bin(3))
print(bin(4))
```

```python
arr2 =np.array([1,2,3,4], dtype='int8')
arr2
```

### Cambiare il tipo del Array

Possiamo cambiare il tipo di dato a Float per esempio



```python
arr = np.array([1,2,3,4], dtype='float64')
arr.dtype
```

```python
arr
```

```python
arr = np.array([0,1,2,3,4])
arr.dtype
```

```python
arr = arr.astype(np.bool_)
arr
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.string_)
arr
```

```python
arr = np.array(['15','1','2','3','4'])
arr = arr.astype(np.int8)
arr
```
