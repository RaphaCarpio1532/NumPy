## Importare librerie

```python
import numpy as np
```

### Arange

```python
list(range(0,10)) #Questo codice non √® di numpy, √® di python.
```

```python
np.arange(0,20,2) #Questo √® un array che va dal 0 al 20 a passi di due
```

### Zeros

```python
np.zeros((10,5)) #Permette di definire una struttura previa (array di tipo matrice)
                  #fatto chiaramente con valori di 0
```

### Ones

```python
np.ones((10,15))
```

### Linspace

```python
np.linspace(0,10,100) #Con linspace posso specificare, da dove inizio 0, dove finisco 10,
                      #quanti valori voglio generare
```

### Eye

```python
np.eye(4) #Crea una matrice con la diagonale principale composta da uni,
#molto utile in alcuni metodi numerici per la risoluzione di problemi reali, sia lineari che non lineari.
```

### Random

```python
np.random.rand() #Genera un valore casuale tra 0 e 1
```

```python
np.random.rand(4) #Array con valori casuali
```

```python
np.random.rand(4,4) #Array di due dimensioni con valori casuali
```

```python
np.random.randint(1,15) #Crea un intero casuale tra 1 e 15
```

```python
np.random.randint(1,100,(10,10)) #Crea una tupla dal 1 al 100 ma di dieci in dieci
```
